```python
#Get Averaged Values

import requests

url = "https://api.openaq.org/v2/averages?date_from=2000-01-01T00%3A00%3A00%2B00%3A00&date_to=2022-09-29T15%3A10%3A00%2B00%3A00&limit=100&page=1&offset=0&sort=desc&group=false"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)
average_values = response.text
print(response.text)
```

```python
#Get Cities

import requests

url = "https://api.openaq.org/v2/cities?limit=100&page=1&offset=0&sort=asc&order_by=city"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

print(response.text)
```

```python
#Get Countries

import requests

url = "https://api.openaq.org/v2/countries?limit=200&page=1&offset=0&sort=asc&order_by=country"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

print(response.text)
```
